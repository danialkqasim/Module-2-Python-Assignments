## 4.1

```python
secret = 7  # any number between 1 and 10
guess = 5   # any number between 1 and 10

if guess < secret:
    print('too low')
elif guess > secret:
    print('too high')
else:
    print('just right')
```

## 6.2

```python
guess_me = 7
number = 1

while True:
    if number < guess_me:
        print('too low')
    elif number == guess_me:
        print('found it!')
        break
    else:
        print('oops')
        break
    number += 1
```

## 6.3

```python
guess_me = 5

for number in range(10):
    if number < guess_me:
        print('too low')
    elif number == guess_me:
        print('found it!')
        break
    else:
        print('oops')
        break
```
